```python
import json
from biolink_to_text import paths_from_trapi, to_text, path_to_text, paths_to_text
```

```python
with open("example_data/ROBOKOP_answer.json", "r") as f:
    trapi_message = json.load(f)["message"]

flat_trapi_paths = paths_from_trapi(trapi_message)
flat_path_texts = paths_to_text(flat_trapi_paths)

enum_trapi_paths = paths_from_trapi(trapi_message, enumerate_edges=True)
enum_path_texts = paths_to_text(enum_trapi_paths)

grouped_trapi_paths = paths_from_trapi(trapi_message, enumerate_edges=True, group_by_result=True)
grouped_path_texts = paths_to_text(grouped_trapi_paths)
```
